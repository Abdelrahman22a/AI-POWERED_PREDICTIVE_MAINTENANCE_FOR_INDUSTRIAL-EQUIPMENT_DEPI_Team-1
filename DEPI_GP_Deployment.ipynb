```python
from flask import Flask, request, jsonify
import joblib
import numpy as np

app = Flask(__name__)
with open("C:/Users/EL-DEWALY/OneDrive/Desktop/DEPI/Deployment/random_forest_model_05.pkl", "rb") as f:
    model = joblib.load(f)

with open("C:/Users/EL-DEWALY/OneDrive/Desktop/DEPI/Deployment/scaler_1.pkl", "rb") as f:
    scaler = joblib.load(f)

@app.route("/")
def home():
    return "FD004 Random Forest model is live!"

@app.route("/predict", methods=["POST"])
def predict():
    try:
        data = request.get_json()

        features = np.array(data["features"]).reshape(1, -1)

        # Apply scaling
        features_scaled = scaler.transform(features)

        prediction = model.predict(features_scaled)[0]
        return jsonify({"predicted_RUL": float(prediction)})


    except Exception as e:
        return jsonify({"error": str(e)})
if __name__ == "__main__":
    app.run()
```
